```python
import matplotlib.pyplot as plt
import numpy as np

import matplotlib as mpl
```

```python
# python
# Data extracted from the LaTeX table
instances = ['1C512M', '1C1G', '2C2G', '4C4G', '8C8G', '8C16G', 'cluster']

# Throughput data
sync_run_throughput = [199.84, 208.46, 321.75, 378.79, 449.24, 504.80, 280.20]
sync_load_throughput = [143.91, 155.59, 328.19, 342.58, 312.70, 333.22, 269.25]
async_run_throughput = [334.78, 280.90, 580.72, 706.71, 835.42, 887.31, 390.47]
async_load_throughput = [283.21, 280.50, 381.10, 503.02, 606.43, 655.74, 374.53]

# Latency data
sync_read_latency = [1926.20, 2200.00, 1350.53, 1577.96, 1167.11, 1164.48, 2065.86]
sync_insert_latency = [3040.31, 2759.08, 1533.21, 1911.53, 1459.78, 1345.03, 2686.16]
sync_update_latency = [2748.89, 2352.66, 1290.78, 1745.94, 1415.80, 1261.76, 3027.99]
async_read_latency = [1547.13, 869.57, 1023.35, 627.73, 794.75, 717.88, 1852.42]
async_insert_latency = [2170.49, 2148.06, 1615.62, 1364.08, 970.46, 1009.33, 2176.81]
async_update_latency = [2367.36, 3211.75, 1318.16, 1230.50, 921.25, 865.08, 2157.16]
```

```python
import matplotlib.pyplot as plt

# Function to plot throughput chart
def plot_throughput():
    plt.figure(figsize=(10, 5))
    plt.plot(instances, sync_run_throughput, label='Sync RunThroughput', marker='o')
    plt.plot(instances, sync_load_throughput, label='Sync LoadThroughput', marker='s')
    plt.plot(instances, async_run_throughput, label='Async RunThroughput', marker='^')
    plt.plot(instances, async_load_throughput, label='Async LoadThroughput', marker='*')
    
    plt.title('Throughput vs Instances')
    plt.xlabel('Instance')
    plt.ylabel('Throughput (ops/sec)')
    plt.legend()
    plt.grid(True)
    plt.show()

# Function to plot latency chart
def plot_latency():
    plt.figure(figsize=(10, 5))
    plt.plot(instances, sync_read_latency, label='Sync ReadLatency', marker='o')
    plt.plot(instances, sync_insert_latency, label='Sync InsertLatency', marker='s')
    plt.plot(instances, sync_update_latency, label='Sync UpdateLatency', marker='^')
    plt.plot(instances, async_read_latency, label='Async ReadLatency', marker='*')
    plt.plot(instances, async_insert_latency, label='Async InsertLatency', marker='p')
    plt.plot(instances, async_update_latency, label='Async UpdateLatency', marker='x')

    plt.title('Latency vs Instances')
    plt.xlabel('Instance')
    plt.ylabel('Latency (us)')
    plt.legend()
    plt.grid(True)
    plt.show()

# Plot the charts
plot_throughput()
plot_latency()
```
